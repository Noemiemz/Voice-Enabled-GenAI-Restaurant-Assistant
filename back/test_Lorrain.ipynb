```python
from transformers import pipeline

pipe = pipeline("text-to-speech", model="maya-research/maya1")
```

```python
from snac import SNAC
import soundfile as sf

description = "Realistic male voice in the 30s age with American accent. Normal pitch, warm timbre, conversational pacing."
text = "Hello! This is Maya1 <laugh_harder> the best open source voice AI model with emotions."

prompt = f'<description="{description}"> {text}'

# Generate speech
output = pipe(prompt)

print("Saving audio to output.wav")
print(output)
# Save the audio to a file

audio_array = output['audio']
sampling_rate = 24000  # Maya1 uses 24 kHz

# Save as WAV
sf.write("output.wav", audio_array.T, sampling_rate)  
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained("maya-research/maya1", torch_dtype=torch.bfloat16, device_map="auto", trust_remote_code=True).to("cuda")
tokenizer = AutoTokenizer.from_pretrained("maya-research/maya1", trust_remote_code=True)
snac_model = SNAC.from_pretrained("hubertsiuzdak/snac_24khz").eval().to("cuda")

description = "Female, in her 30s with an American accent and is an event host, energetic, clear diction"
text = "Wow. This place looks even better than I imagined. How did they set all this up so perfectly? The lights, the music, everything feels magical. I can't stop smiling right now."

prompt = f'<description="{description}"> {text}'
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

print("Generating audio...")
with torch.no_grad():
    outputs = model.generate(**inputs, max_new_tokens=2048, temperature=0.4, top_p=0.9, repetition_penalty=1.1, do_sample=True)

# Extract and decode audio (see full example in the docs)
audio_tokens = outputs[:, inputs["input_ids"].shape[1]:]

# Decode audio tokens and ensure the correct dimensions
audio = snac_model.decode(audio_tokens).squeeze(0).cpu().numpy()


# Save the audio to a file
sf.write("output.wav", audio, 24000)
```
