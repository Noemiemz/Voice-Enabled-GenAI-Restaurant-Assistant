### Comment enregistrer un audio √† partir d'un code ?

```python
import sounddevice as sd
import numpy as np
import wavio
```

```python
sample_rate = 44100
channels = 1

# contiendra les morceaux de l‚Äôaudio
audio_chunks = []
stream = None

def audio_callback(indata, frames, time, status):
    """Cette fonction est appel√©e automatiquement √† chaque chunk d‚Äôaudio"""
    if status:
        print(status)
    audio_chunks.append(indata.copy())  # on stocke le morceau

def start_recording():
    global stream, audio_chunks

    audio_chunks = []  # reset
    print("üéôÔ∏è Enregistrement d√©marr√©...")

    # ouverture d‚Äôun stream audio
    stream = sd.InputStream(
        samplerate=sample_rate,
        channels=channels,
        callback=audio_callback
    )

    stream.start()

def stop_recording():
    global stream, audio_chunks

    print("‚èπÔ∏è Enregistrement arr√™t√©.")
    stream.stop()
    stream.close()

    # concat√®ne tous les morceaux
    audio_data = np.concatenate(audio_chunks, axis=0)

    # sauvegarde propre
    wavio.write("audio/audio.wav", audio_data, sample_rate, sampwidth=2)
    print("üíæ Fichier sauvegard√© : audio.wav")
```

```python
input("Appuie sur Entr√©e pour D√âMARRER l‚Äôenregistrement")
start_recording()

input("Appuie sur Entr√©e pour ARR√äTER l‚Äôenregistrement")
stop_recording()
```
