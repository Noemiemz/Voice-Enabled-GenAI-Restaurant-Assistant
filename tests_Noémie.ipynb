### Comment enregistrer un audio √† partir d'un code ?

```python
import sounddevice as sd
import numpy as np
import wavio
```

recording = None          # contiendra les donn√©es audio
sample_rate = 44100       # qualit√© audio (CD quality)
duration_limit = 120      # limite max (sec) si on oublie d'arr√™ter

def start_recording():
    global recording

    print("üéôÔ∏è Enregistrement lanc√©...")

    # On RECOMMANDE de d√©finir une dur√©e max pour ne pas remplir la RAM.
    recording = sd.rec(
        int(duration_limit * sample_rate),
        samplerate=sample_rate,
        channels=1
    )

def stop_recording():
    global recording

    print("‚èπÔ∏è Arr√™t de l‚Äôenregistrement...")

    # arr√™te imm√©diatement la capture audio
    sd.stop()

    # convertit les donn√©es en tableau numpy
    audio = np.array(recording)

    # enregistre dans un fichier wav
    wavio.write("audio.wav", audio, sample_rate, sampwidth=2)

    print("üíæ Fichier sauvegard√© : audio.wav")
